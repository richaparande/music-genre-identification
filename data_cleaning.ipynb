```python
import numpy as np
import pandas as pd
import utils
```

```python
# load echonest data using the utils.py file from the FMA GitHub repository

echonest = utils.load('fma-master/data/fma_metadata/echonest.csv')
echonest.head()
```

```python
echonest.columns
```

```python
# drop the top level of columns to gain access to the track audio features

echonest.columns = echonest.columns.droplevel()
audio_features = echonest['audio_features']
audio_features.head()
```

```python
# save audio features to a csv file for analysis

audio_features.to_csv('audio_features.csv')
```

```python
# make sure the export worked

audio_features = pd.read_csv('audio_features.csv', index_col='track_id')
audio_features.head()
```

```python
# load all track data using utils.py

all_tracks = utils.load('fma-master/data/fma_metadata/tracks.csv')
all_tracks.head()
```

```python
all_tracks.columns
```

```python
# drop the top level of tracks to gain access to the genre columns

all_tracks.columns = all_tracks.columns.droplevel()
track_genres = all_tracks[['genre_top', 'genres', 'genres_all']].copy()
track_genres.head()
```

```python
# load genre data using utils.py

genres = utils.load('fma-master/data/fma_metadata/genres.csv')
genres.head()
```

```python
# select only the tracks that have corresponding audio features, as those are the only ones I can use in my analysis

echonest_tracks = track_genres.loc[track_genres.index.isin(audio_features.index.values)]
echonest_tracks.head()
```

```python
# take a look at the size of the dataset

echonest_tracks.shape
```

```python
# check if there are any missing top genres

echonest_tracks['genre_top'].isnull().values.any()
```

```python
# separate the tracks that have missing genres

missing_genre = echonest_tracks[echonest_tracks.isna().any(axis=1)]
missing_genre.head()
```

```python
len(missing_genre)
```

```python
# create a genre dictionary to match the genres

genre_dict = dict(zip(genres.index, genres.title))
genre_dict
```

```python
# create a list of genre ids for each track

missing_list = missing_genre.genres.tolist()
missing_list[:10]
```

```python
# replace genre ids with genre titles using the dictionary

for lst in missing_list:
    for x, item in enumerate(lst):
        lst[x] = genre_dict.get(item, item)
missing_list[:10]

# source: https://codereview.stackexchange.com/questions/79891/iterate-over-large-list-of-lists-and-replace-its-elements
# full reference in bibliography
```

```python
# use only the first genre in each list, as I assume that to be the primary genre

missing_lst = [x[0] for x in missing_list]
missing_lst[:10]
```

```python
# create a dataframe of track ids and new genre labels

track_id = list(missing_genre.index)
missing = pd.DataFrame(missing_lst, index = track_id, columns = ['genre_top'])
missing.head()
```

```python
# add genre columns

missing['genres'] = missing_genre['genres']
missing['genres_all'] = missing_genre['genres_all']
missing.head()
```

```python
# merge missing data with echonest tracks

merging = pd.concat([missing, echonest_tracks])
merging.head()
```

```python
merging = merging.dropna()
```

```python
len(merging)
```

```python
# check again to see if there are any missing genres

merging['genre_top'].isnull().values.any()
```

```python
# merge genre data with audio features

merged = audio_features.join(merging)
merged.head()
```

```python
# save merged tracks to a csv file for analysis

merged.to_csv('merged.csv')
```
